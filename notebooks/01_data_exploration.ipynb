```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 50)
print("Libraries loaded sucessfully")
```

```python
#load dataset
df = pd.read_csv("../data/diabetic_data.csv")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
#Preview data
df.head(10)
```

```python
#check datatype
df.dtypes
```

```python
# Find missing values (the ? problem)
#replace all "?", with NaN first
df.replace("?", np.nan, inplace=True)

#Now count missing values per column
missing = df.isnull().sum()
missing_pct = (missing / len(df) *100).round(1)

# create a summary table of missing values
missing_summary = pd.DataFrame({
    "missing_count" : missing,
    "missing_pct"   : missing_pct
}).query("missing_count > 0").sort_values( "missing_pct", ascending=False )

print(missing_summary )
```

```python
#Understanding target variable
target_counts = df["readmitted"].value_counts()
target_pct   = df["readmitted"].value_counts(normalize=True).mul(100).round(1)

#create a summary table for target variable
target_summary = pd.DataFrame({
    "Count"       : target_counts,
    "Percentage"  : target_pct
    })
print(target_summary)
```

```python
#cleaning all the missing values
#Drop columns with too much missing data
cols_to_drop = ["weight", "max_glu_serum", "medical_specialty", "payer_code"]
df.drop(columns=cols_to_drop, inplace=True)
print(f"Columns remaining: {df.shape[1]}") 
```

```python
#Handling A1Cresult(turn missing into feature)
#Missing A1Cresult means the test was never ordered 
#This is actually a meaningful clinical info-create a binary feature 
df["A1C_tested"] = df["A1Cresult"].apply(lambda x: 0 if pd.isna(x) else 1)


#For the original column, fill missing with "Not tested"
df["A1Cresult"] = df["A1Cresult"].fillna("Not test")

print(df["A1C_tested"].value_counts())
print(df["A1Cresult"].value_counts())
                                
```

```python
#Handling race and diagnosis column
#Fill missing race with most frequent value
df["race"] = df["race"].fillna(df["race"].mode()[0])

#Drop the small number of rows missing diagnosis codes
df.dropna(subset=["diag_1", "diag_2", "diag_3"], inplace=True)

print(f"Rows after  dropping missing diagnoses: {len(df)}")
```

```python
#Handling duplicate patients
#Dataset has multiple visits per patient-keep only first encounter
#This prevents data leakage(a patient's later visits infleuncing earlier predictions)
df = df.sort_values("time_in_hospital", ascending=False)
df = df.drop_duplicates(subset="patient_nbr", keep="first")

print(f"Rows after removing duplicate patients: {len(df)}")
```

```python
#Creating he binary target variable
#Convert the readmitted to binary : 1 = readmitted within 30 days, 0 = everytthing else
df["readmitted_30"] = df["readmitted"].apply(lambda x: 1 if x == "<30" else 0)

# check the class balance
counts = df["readmitted_30"].value_counts()
pcts   = df["readmitted_30"].value_counts(normalize=True).mul(100).round(1)

print(pd.DataFrame({"counts" : counts, "percentage" : pcts}))
```

```python
# confirming the claen dataset
print(f"Final dataset shape: {df.shape}")
print(f"\nAny remaining missing values?")
print(df.isnull().sum()[df.isnull().sum()>0])

#Save the clean data version for use in the next phase
df.to_csv("../data/diabetic_data_clean.csv", index=False)
```

```python
#Encode the target and remove irrevant ccolumns
#Drop columns that are IDs or leak feature informations
cols_to_remove = ["encounter_id", "patient_nbr", "readmitted"]
df.drop(columns=cols_to_remove, inplace=True)

print(f"Shape: {df.shape}")
```

```python
#Convert age brackets to numbers
#Age comes as brackets like [50-60), convert to ordinal numbers
age_map = {
    "[0-10)":1, "[10-20)":2,  "[20-30)" :3, "[30-40)" :4, "[40-50)":5, 
    "[50-60)":6, "[60-70)":7, "[70-80)":8, "[80-90)":9, "[90-100)":1
}

df["age_num"] = df["age"].map(age_map)
df.drop(columns=["age"], inplace=True)

print(df["age_num"].value_counts().sort_index())
```

```python
#Creating clinical risk features

#Total procedures and medications as a complexity proxy
df["total_interactions"] = df["num_medications"] + df["num_procedures"] + df["num_lab_procedures"]

#Prior hospital usage -strong readmission predictions
df["prior_hospital_use"] = df["number_outpatient"] + df["number_emergency"] + df["number_inpatient"]

#flag long stays (7+ days often indcates severity)
df["long_stay"] = (df["time_in_hospital"] >=7).astype(int)

print(df[["total_interactions", "prior_hospital_use", "long_stay"]].describe())
```

```python
#simplifying 23 medication columns
#Each drug column has values No, steady, Up, Down
#Encode: 0=not prescribed, 1 = prescribed(any dose)
med_cols = [
    "metformin", "repaglinide", "nateglinide", "chlorpropamide",
            "glimepiride", "acetohexamide", "glipizide", "glyburide",
            "tolbutamide", "pioglitazone", "rosiglitazone", "acarbose",
            "miglitol", "troglitazone", "tolazamide", "examide",
            "citoglipton", "insulin", "glyburide-metformin",
            "glipizide-metformin", "glimepiride-pioglitazone",
            "metformin-rosiglitazone", "metformin-pioglitazone"
]

for col in med_cols:
    df[col] = df[col].apply(lambda x : 0 if  x == "No" else 1)

#Also create a total meds count feature
df["total_meds_prescribed"] = df[med_cols].sum(axis=1)

print(f"Avg medications prescribed: {df["total_meds_prescribed"].mean():.1f}")
```

```python
#Encode remaining categorical columns
from sklearn.preprocessing import LabelEncoder

cat_cols = ["race", "gender", "A1Cresult", "change", "diabetesMed"]

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

print("Categorical columns encoded.")
print(df[cat_cols].head())
```

```python
#droping diagnosis columns for now
#diag_1/2/3 are ICD _9 codes with 700+ unique values
#Grouping them properly takes a fulll notebook on its own
#For v1 of this project, droping this the best option now

df.drop(columns=["diag_1", "diag_2", "diag_3"], inplace=True)

print(f"Final shape: {df.shape}")
print(f"\nAll columns:\n{df.columns}")
```

```python
# Check where the typo came from
print(df.columns[df.columns.str.contains('meds')].tolist())

# Drop the typo column
df.drop(columns=['total_meds_ prescribed'], inplace=True)

# Confirm it's gone
print(f"Fixed. Shape: {df.shape}")
print(df.columns.tolist())
```

```python
#Saving and do final checks
#confirm no missing values remain
print("Missing values remaining:")
print(df.isnull().sum()[df.isnull().sum()>0])

# save for modelling notebook
df.to_csv("../data/diabetic_data_features.csv", index=False)
print(f"\nFeature dataset saved: shape: {df.shape}")
```
